```python
import gzip
import matplotlib.pyplot as plt

def readDraw(f, j=0):
    with gzip.open(f, 'rt', encoding='utf-8') as file:
        i = 0
        steps = []
        globPpx = []
        bucket0 = []
        bucket1 = []
        bucket2 = []
        bucket3 = []
        bucketMean = []

        for line in file:
            i = i + 1
            l = line.split()

            if "global" in l:
                steps.append(l[2])
                globPpx.append(l[-1])
            if "bucket" in l:
                exec('bucket' + l[2] + '.append(' + l[-1] + ")")
            if "mean" in l:
                bucketMean.append(l[-1])

        if len(steps) == len(bucket0) and len(bucket3) == len(bucketMean):
            print("complete")

    plt.xlabel("steps")
    plt.ylabel("perplexity")
    plt.plot(steps[j:], globPpx[j:], label='Train Perplexity')
    plt.plot(steps[j:], bucket0[j:], label='Validation: bucket 0')
    plt.plot(steps[j:], bucket1[j:], label='Validation: bucket 1')
    plt.plot(steps[j:], bucket2[j:], label='Validation: bucket 2')
    plt.plot(steps[j:], bucket3[j:], label='Validation: bucket 3')

    plt.legend(loc=1)

    plt.title(str(f))
    plt.show()

# Define the path to your compressed dataset file (e.g., .gz file)
file_path = "/Users/ISHAN/Downloads/multi30k_test_sets_d3ec2a38.txt"

# Call the readDraw function with your dataset file
readDraw(file_path)
```
